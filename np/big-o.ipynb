```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Define the range for n
n = np.linspace(1, 1000, 400)

# Define the functions for O(n), O(log n), and O(n log n)
O_n = n
O_log_n = np.log2(n)
O_n_log_n = n * np.log2(n)
O_2n_log_n = 2 * n * np.log2(n)
O_n2 = n ** 2

# Plot the functions
plt.figure(figsize=(10, 6))
plt.plot(n, O_n, label='O(n)', color='blue')
plt.plot(n, O_log_n, label='O(log n)', color='green')
plt.plot(n, O_n_log_n, label='O(n log n)', color='red')
plt.plot(n, O_n2, label='O(n ^ 2)', color='yellow')

# Add labels and legend
plt.xlabel('Input Size (n)')
plt.ylabel('Operations')
plt.title('Complexity Comparison: O(n), O(log n), O(n log n)')
plt.legend()
plt.grid(True)
plt.yscale('log')  # Use logarithmic scale for better visibility

# Show the plot
plt.show()
```
